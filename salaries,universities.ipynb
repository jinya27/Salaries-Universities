```python
plt.plot(df['SAT'],df['Accept'], 'go') #go represents shape nd color
plt.xlabel('SAT')
plt.ylabel('Accept')
plt.title('Scatterplot')
plt.show()
```

```python
plt.plot(df['SAT'],df['Top10'], 'rs')
plt.xlabel('SAT')
plt.ylabel('Top10')
plt.title('Scatterplot')
plt.show()
```
